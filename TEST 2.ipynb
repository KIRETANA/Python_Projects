## QUESTION 1

```python
import scipy as sc
import numpy as np
import matplotlib.pyplot as plt

#importing all libraries required

y0 = 1 #initial condition

start = 0
end = 2
t = np.linspace(start,end,end) #setting up the interval for t 

def dydt(y,t):
    #the differential equation
    dydt = -2 * y * t
    return dydt

y_initial = np.linspace(y0,y0,end)

y = sc.integrate.odeint(dydt,y0,t) #integration of the differential quation

print('The answer for the ODE is:',y)

#plotting the functions
plt.figure()
plt.plot(t, y,'r','change in y value')
plt.plot(t,y_initial, 'b--',linewidth=2)
plt.title('integration og dydt = -2yt')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()
plt.show()
```
